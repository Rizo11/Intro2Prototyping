# **Final project**
As a final project, you will need to assemble a parallel type shoulder for robot arm with 2 rotational dof. You will need to do all the preliminary calculations, set up numerical experiments and find out the maximum external load, develop a 3d model and set up the simplest control.

You can find the shoulder mechanisms in papers:
* [A Comparison of Robot Wrist Implementations for the iCub Humanoid](https://www.researchgate.net/publication/331168570_A_Comparison_of_Robot_Wrist_Implementations_for_the_iCub_Humanoid)

* [Kinematics and dynamics analysis of a 2-DOF spherical parallel robot](https://ieeexplore.ieee.org/abstract/document/7886838)

* [Kinematic Synthesis and Modeling of a Three Degrees-of-Freedom Hybrid Mechanism for Shoulder and Hip Modules of Humanoid Robots](https://asmedigitalcollection.asme.org/mechanismsrobotics/article-abstract/8/4/041017/384194/Kinematic-Synthesis-and-Modeling-of-a-Three)

* [Development of a cybernetic shoulder-a 3-DOF mechanism that imitates biological shoulder motion](https://ieeexplore.ieee.org/abstract/document/1435487)

Or you can suggest your own shoulder mechanism. The main thing is that it should be parallel and 2 dof.
